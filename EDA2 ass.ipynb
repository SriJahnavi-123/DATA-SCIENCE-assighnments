```python
le = LabelEncoder()
dumm=df.copy()

dumm['workclass'] = le.fit_transform(dumm['workclass'])
dumm['education'] = le.fit_transform(dumm['education'])
dumm['marital_status'] = le.fit_transform(dumm['marital_status'])
dumm['occupation'] = le.fit_transform(dumm['occupation'])
dumm['relationship'] = le.fit_transform(dumm['relationship'])
dumm['race'] = le.fit_transform(dumm['race'])
dumm['sex'] = le.fit_transform(dumm['sex'])
dumm['native_country'] = le.fit_transform(dumm['native_country'])
```

```python
# dumm=df.copy()
dumm = pd.get_dummies(dumm, columns=['income'])
dumm
```

```python
clf= IsolationForest(random_state=10,contamination=.01)
clf.fit(dumm)
```

```python
y_pred_outliers = clf.predict(dumm)
```

```python
y_pred_outliers
```

```python
pd.Series(y_pred_outliers).value_counts()
```

```python
dumm['anomaly']=clf.predict(dumm.iloc[:,0:16])
dumm
```
