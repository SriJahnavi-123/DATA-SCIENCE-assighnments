```python
#Apriori algorithm
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
!pip install mlxtend
```

```python
transcation = []
for i in range(data1.shape[0]):
    transcation.append([str(data1.values[i,j]) for j in range(0,data1.shape[1])])
```

```python
te = TransactionEncoder()
te_data = te.fit(transcation).transform(transcation)
```

```python
online_retail = pd.DataFrame(te_data,columns=te.columns_)
```

```python
online_retail
```

```python
freq_itemset = apriori(online_retail,min_support=0.1,use_colnames=True)
```

```python
freq_itemset
```

```python
rules = association_rules(freq_itemset,metric='lift',min_threshold=1)
rules.sort_values('lift',ascending=False).head(10)
rules
```

```python
# Experiment with different values
min_support = 0.03
min_confidence = 0.5
min_lift = 3

freq_itemset = apriori(online_retail, min_support=min_support, use_colnames=True)
rules = association_rules(freq_itemset, metric='lift', min_threshold=min_lift)
rules = rules[rules['confidence'] >= min_confidence]
```

```python
rules
```

```python
plt.bar(rules['antecedents'].apply(lambda x: list(x)[0]), rules['confidence'])
plt.xlabel('Antecedent')
plt.ylabel('Confidence')
plt.title('Confidence by Antecedent Bar Chart')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.scatter(rules['support'], rules['confidence'])
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.title('Support vs Confidence Scatter Plot')
plt.show()
```

**1.	What is lift and why is it important in Association rules?**

Lift in association rule mining measures how much more likely two items are to be bought together than if they were purchased independently.
Identifies True Associations: A lift greater than 1 indicates a positive association, meaning the items are more likely to be purchased together. A lift of 1 suggests no association (they are independent), and a lift less than 1 implies a negative association (they are less likely to be bought together).


**2.	What is support and Confidence. How do you calculate them?**

**Support:**Support measures how frequently an itemset (a set of items) appears in the dataset. It represents the popularity of an itemset


**Confidence:**Confidence measures the strength of the association between the antecedent (the "if" part) and the consequent (the "then" part) of a rule. It indicates how likely the consequent is to be purchased when the antecedent is already in the basket.
